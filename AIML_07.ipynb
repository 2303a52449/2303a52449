```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
df.describe()
```

```python
np.sum(pd.isnull(df).sum())
```

```python
cancer['target']
```

```python
df['cancer']=pd.DataFrame(cancer['target'])
df.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='cancer',data=df,palette='RdBu_r')
```

```python
l=list(df.columns[0:10])
for i in range(len(l)-1):
  sns.boxplot(x='cancer',y=l[i],data=df,palette='rainbow')
  plt.show()
```

```python
f,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(12,10))
ax1.scatter(df['mean area'],df['cancer'])
ax1.set_title("cancer cases as a function of mean area",fontsize=15)
ax2.scatter(df['mean smoothness'],df['cancer'])
ax2.set_title("cancer cases as a function of mean perimeter ")
```
